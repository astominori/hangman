```python
def hangman(word):
	wrong = 0
	stages = ["",
	         "_________       ",
	         "|               ",
	         "|        |      ",
	         "|        O      ",
	         "|       /|\     ",
	         "|       / \     ",
	         "|               ",
	         ]
	rletters = list(word)
	board = ["_"] * len(word)
	win = False
	print('ハングマンへようこそ！')
	count = 1
	
	while wrong < len(stages) - 1:
		print('\n')
		msg = (str(count)+'文字目を予想してね:')
		char = input(msg)
		flag = char in rletters
		if flag == True:
			cind = rletters.index(char)
			if cind == count-1:
					board[cind] = char
					rletters[cind] = '$'
					print('正解！{}文字目は{}'.format(count,char))
					count += 1
			else:
					print('おしい！{}は文字として含まれています'.format(char))
					wrong += 1
		else:
			wrong += 1
		print(' '.join(board))
		e = wrong +1
		print('\n'.join(stages[0:e]))
		if "_" not in board:
			print('あなたのかち！')
			print(' '.join(board))
			win = True
			break
	if not win:
		print('\n'.join(stages[0:wrong+1]))
		print('あなたの負け！正解は{}'.format(word))
	
word = input('\n好きな英単語を入れてね:')
hangman(word)

```
